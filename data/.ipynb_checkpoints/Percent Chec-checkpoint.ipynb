```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

fifth = pd.read_csv('final_predictions_5.csv')
last = pd.read_csv('../final_predictions_last.csv')
tweets = pd.read_excel('final_tweets.xlsx')
```

```python
#first = first.drop(['Unnamed: 0'], axis = 1)
#second = second.drop(['Unnamed: 0'], axis = 1)
#third = third.drop(['Unnamed: 0'], axis = 1)
#fourth = fourth.drop(['Unnamed: 0'], axis = 1)
fifth = fifth.drop(['Unnamed: 0'], axis = 1)
last = last.drop(['Unnamed: 0'], axis = 1)
tweets = tweets.drop(['Unnamed: 0'], axis = 1)
```

```python
def find_percent(col, name):
    count = len(pd.Series(col)[col == 'bot'])
    length = len(pd.Series(col)[col == 'bot']) + len(pd.Series(col)[col == 'human'])
    return count / length

for (first_name, first_col), (fifth_name, fifth_col) in zip(last.iteritems(), fifth.iteritems()):
    print('first', first_name + ":", find_percent(first_col, first_name))
    print('fifth', fifth_name + ":", find_percent(fifth_col, fifth_name))
```

```python
for ((first_name, first_col), (tweet_name, tweet_col)) in zip(last.iteritems(), tweets.iteritems()):
    length1 = len(pd.Series(first_col)[first_col == 'bot']) + len(pd.Series(first_col)[first_col == 'human'])
    length2 = len(pd.Series(tweet_col).dropna())
    print(first_name, length1, length2)
```

```python
results = []
new_res = pd.DataFrame()
for name, col in fifth.iteritems():
    if len(pd.Series(col)[col == 'bot']) + len(pd.Series(col)[col == 'human']) > 20:
        new_res[name] = col
for name, column in new_res.iteritems():
    results.append(find_percent(column, name))

plt.barh(y = new_res.columns, width = results, linewidth = 50)
plt.rcParams.update({'font.size': 10})
plt.title('Bot Percentages for Developments with 20+ Tweets')
plt.show()
```
